```python
# Add labels to the x- and y-axis:

import numpy as np
import matplotlib.pyplot as plt

x = np.array([11, 15, 20, 24, 27, 30])
y = np.array([4, 4, 5.5, 6, 6, 6.2])

plt.plot(x, y)

plt.xlabel("Age")
plt.ylabel("Height")

plt.show()
```

```python
# Add a plot title and labels for the x- and y-axis:
import numpy as np
import matplotlib.pyplot as plt

x = np.array([11, 15, 20, 24, 27, 30])
y = np.array([4, 4, 5.5, 6, 6, 6.2])

plt.plot(x,y)

plt.title("Student Age & Height")
plt.xlabel("Age")
plt.ylabel("Height")

plt.show()
```

```python
# Set font properties for the title and labels:
import numpy as np
import matplotlib.pyplot as plt

x = np.array([11, 15, 20, 24, 27, 30])
y = np.array([4, 4, 5.5, 6, 6, 6.2])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}


plt.title("Student Age & Height",fontdict = font1)
plt.xlabel("Age",fontdict = font2)
plt.ylabel("Height",fontdict = font2)


plt.plot(x,y)
plt.show()
```

```python
# Position the title to the left:
import numpy as np
import matplotlib.pyplot as plt

x = np.array([11, 15, 20, 24, 27, 30])
y = np.array([4, 4, 5.5, 6, 6, 6.2])


plt.title("Student Age & Height",loc = 'left')
plt.xlabel("Age")
plt.ylabel("Height")


plt.plot(x,y)
plt.show()
```
