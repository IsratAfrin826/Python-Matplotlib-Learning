```python
# Draw three plots on the same figure:

import matplotlib.pyplot as plt
import numpy as np

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

#day three, the age and speed of 17 cars:
x = np.array([3,3,8,10,13,16,17,19,22,25,27,28,33,34,23,44,40])
y = np.array([80,56,45,46,37,100,130,123,45,78,98,66,78,90,95,99,100])
plt.scatter(x, y)

plt.show()
```

```python
# Include the actual colormap:

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6,9,10,17,23,44,67,87,90,80,45,78,34])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86,123,124,161,178,99,78,56,59,69,77,88,99])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100,110,120,130,140,150,160,170,180,190,200,210,220])

plt.scatter(x, y, c=colors, cmap='rainbow')

plt.colorbar()

plt.show()
```

```python
# Create random arrays with 100 values for x-points, y-points, colors and sizes:

import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(150, size=(150))
y = np.random.randint(150, size=(150))
colors = np.random.randint(150, size=(150))
sizes = 10 * np.random.randint(150, size=(150))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```
